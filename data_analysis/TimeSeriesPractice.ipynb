```
# Moving average function
def moving_average(a, n=3):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

Moving average every 1000

```
X100mean3 = moving_average(X100abs, 1000)
y100mean3 = y100[:999001]

t = np.arange(1, 99900100, 100)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time')
ax1.set_ylabel('Acustic Data', color=color)
ax1.plot(t, X100mean3, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Time To Failure', color=color)  # we already handled the x-label with ax1
ax2.plot(t, y100mean3, color=color)
ax2.tick_params(axis='y', labelcolor=color)
```
